# Monty Hall Solucion Aberg Cobo

```python
import random


gano_quedandome = 0
gano_cambiandome = 0


#Genero las listas para el grafico

gano_quedandome_acum = []
gano_cambiandome_acum = []



#Repito el experimento 100.000 veces
n = 100000

for repeticiones in range (n):

#Mi seleccion es al azar entre las 3 puertas, y el premio es escondido al azar 
#en alguna de las 3 puertas.
    
    puertas = [1,2,3]
    seleccion = random.randint(1,3)
    premio = random.randint(1,3)
    
#Despues de haber seleccionado, una de las puertas, el conductor selecciona sobre las dos restantes.
#Por lo tanto, elimino mi seleccion.

    puertas.remove(seleccion)
    
#El conductor puede elegir entre dos puertas, habria que eliminar la puerta donde esta el premio, ya que
#si mi seleccion no coincide con el premio, logro que el conductor no abra la puerta donde esta el premio.

    if not(seleccion==premio):
        puertas.remove(premio)
    
    conductor_elige = random.choice(puertas)

#Si mi seleccion coincide con el premio, entonces agrego a la lista correspondiente el acierto de quedandome,
#Si el premio esta en la puerta restante, entonces agrego el evento a la lista en la cual gano cambiandome.

    if seleccion==premio:
        gano_quedandome +=1
    
    else:
        gano_cambiandome +=1
    
    gano_quedandome_acum.append(gano_quedandome)
    gano_cambiandome_acum.append(gano_cambiandome)
  

print('porcentaje_gano_quedandome: ',gano_quedandome/n)
print('porcentaje_gano_cambiandome: ',gano_cambiandome/n)


```

```python
import matplotlib.pyplot as plt

plt.plot(gano_quedandome_acum, label='No cambio')
plt.plot(gano_cambiandome_acum, label='Cambio')
plt.xlabel('Repeticion')
plt.ylabel('Evento')
plt.legend()
plt.show()
```
